```python
class Bank:
    #static
    bank_name = "Bank of HIT"
    bank_loc = "Haldia"
    bank_contact = "100"
    bank_code = 235234

    def __init__(self, username, contact, age, balance):
        self.username = username
        self.contact = contact
        self.age = age
        self.balance = balance

    #object method
    def change_contact(self, contact):
        self.contact = contact

    #object method
    def display(self):
        print(self.username, self.contact, self.age, self.balance)

    @staticmethod
    def is_valid_amount(amt):
        return amt > 0

    def withdraw_money(self, amt):
        if(self.balance >= amt and self.is_valid_amount(amt)):
            self.balance -= amt
        print("Withdrawn money:", amt)

    def deposit_money(self, amt):
        if self.is_valid_amount(amt):
            self.balance += amt
        print("Deposited money:", amt)

    @classmethod
    def bank_details(self):
        print(self.bank_name,"  code:", self.bank_code)

    @classmethod
    def change_code(self, newcode):
        self.bank_code = newcode



u4 = Bank("ABC", 7566435457, 24, 10000)
print(u4.bank_name, u4.username, u4.contact, u4.age)

u5 = Bank("Gamma", 9786573457, 20, 100)
u5.display()
u5.change_contact(9999999999)
u5.display()
```

```python
u4.withdraw_money(1000)
u4.display()
u4.deposit_money(4000)
u4.display()
```

```python
class Ten:
    def __init__(self, name, roll, tenth_marks):
        self.name = name
        self.roll = roll
        self.tenth_marks = tenth_marks

    def display(self):
        print(self.name, self.roll, self.tenth_marks)




class Twelve(Ten):
    def __init__(self, name, roll, tenth_marks, twelve_marks, stream):
        super().__init__(name, roll, tenth_marks)
        self.twelve_marks = twelve_marks
        self.stream = stream
        
    def display_details(self):
        super().display()
        print(self.stream, self.twelve_marks)




class Bachelors(Twelve):
    def __init__(self, name, roll, tenth_marks, twelve_marks, stream, course, branch, cgpa):
        super().__init__(name, roll, tenth_marks, twelve_marks, stream)
        self.course = course
        self.branch = branch
        self.cgpa = cgpa

    def display_all(self):
        super().display_details()
        print(self.course, self.branch, self.cgpa)
```

```python
st1 = Twelve("John", 102, 90, 80, 'PCMB')
st1.display_details()
```

```python
st3 = Bachelors("Ron", 12, 90, 85, 'PCMB', 'BTech', 'CSE', 8.0)
st3.display_all()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, roll, branch):
        super().__init__(name, age)
        self.roll = roll
        self.branch = branch

class Teacher(Person):
    def __init__(self, name, age, subject, experience):
        super().__init__(name, age)
        self.subject = subject
        self.experience = experience


class Academic(Student):
    def __init__(self, name, age, roll, branch, marks):
        super().__init__(name, age, roll, branch)
        self.marks = marks

class Sports(Student):
    def __init__(self, name, age, roll, branch, sport_name, level):
        super().__init__(name, age, roll, branch)
        self.sport_name = sport_name
        self.level = level

class AllRounderStudent(Academic, Sports):
    def __init__(self, name, age, roll, branch, marks, sport_name, level):
        Academic.__init__(name, age, roll, branch, marks)
        Sports.__init__(name, age, roll, branch, sport_name, level)
```

```python
st1 = AllRounderStudent("John", 20, 100, 'CSE', 87, 'football', 'district')
```
