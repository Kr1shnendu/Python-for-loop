```python
c = input("Enter a character: ")

if c.isupper():
    print("Uppercase letter")
elif c.islower():
    print("Lowercase letter")
elif c.isdigit():
    print("Digit")
else:
    print("Special character")
```

```python
exp = int(input("Enter year of experience: "))
rating = int(input("Enter rating: "))

if exp >= 5:
    if rating >= 4:
        print("20% bonus")
    else:
        print("10% bonus")
else:
    print("No bonus")
```
