```python
def count(num):
    even, odd = 0, 0
    while(num != 0):
        d = num % 10
        num = num // 10
        if(d % 2 == 0):
            even+=1
        else:
            odd+=1
    return even, odd

e, o = count(25342)
print("even:",e)
print("odd:",o)
```

```python
def pal(x):
    num = x
    rev = 0
    while(num != 0):
        d = num % 10
        num = num // 10
        rev = (rev * 10) + d
    if(rev==x):
        return True
    else:
        return False

if(pal(5335)):
    print("Palindrome")
else:
    print("Not palindrome")
```
