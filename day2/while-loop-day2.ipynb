```python
i = 1
while(i<=100):
    num = i
    while(num != 0):
        d = num % 10
        num = num // 10
        if(d==7):
            print(i)
            break
    i+=1
```

```python
x = 675

num = x
rev = 0
while(num != 0):
    d = num % 10
    num = num // 10
    rev = (rev * 10) + d

if(rev > x):
    print("reversed number is greater than the original...")
else:
    print("reversed number is not greater than the original...")

```

```python
sum = 0
num = 237
while True:
    while(num != 0):
        d = num % 10
        num = num // 10
        sum += d
    if(sum<10):
        break
    else:
        num = sum
        sum = 0

print(sum)
```

```python
num = 52346
even, odd = 0, 0
while(num != 0):
    d = num % 10
    num = num // 10
    if(d % 2 == 0):
        even+=1
    else:
        odd+=1

print("No of even digits:", even)
print("No of odd digits:", odd)
```

```python
x = 606

num = x
rev = 0
while(num != 0):
    d = num % 10
    num = num // 10
    rev = (rev * 10) + d

if(rev == x):
    print("Palindrome number")
else:
    print("Non palindrome number")
```

```python
attempt = 3
real_name = "abc"
real_pw = "1234"
name = input("Enter name:")

if(real_name==name):
    while(attempt > 0):
        print("Attempt:", attempt)
        attempt -= 1
        pw = input("Enter password:")
        if(real_pw==pw):
            print("Login successful")
            break

    if(attempt==0):
        print("Three trials ended...")
```

```python
digit = 3
num = 25362433
count = 0

while(num != 0):
    d = num % 10
    num = num // 10
    if(d==digit):
        count+=1

print("Digit apprears", count, "times")
```
