# matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1, 50, 10)
y = np.random.randint(1, 50, 10)
print(x, "\n", y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.plot(x, y, color = 'red', linestyle = 'dotted')
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = 'dotted') # '-', '--', '-.', ':'
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = '-.')
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = '-.', alpha = 0.3)
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = '-.', alpha = 0.7, marker='*')
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = '-.', alpha = 0.7, marker='*', markerfacecolor='y', markersize=20, markeredgecolor='blue')
plt.show()
```

```python
plt.title('Random Chart')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.xticks(x)
plt.yticks(y)
plt.plot(x, y, color = 'red', linestyle = '-.', alpha = 0.7,
        marker='*', markerfacecolor='y', markersize=10, markeredgecolor='blue',
        linewidth=5)
plt.show()
```

```python
x = np.arange(1, 11)
y = np.sin(x)
z = np.cos(x)

print(x, y, z, sep="\n")
```

```python
plt.figure(figsize= (8, 4))
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize= (10, 3))

plt.subplot(1, 2, 1) # row=1, col=1, plot_no=1
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Values")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points2")
plt.ylabel("Values2")
plt.legend()

plt.show()
```

```python
plt.figure(figsize= (6, 6))

plt.subplot(3, 3, 1) # row=1, col=1, plot_no=1
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Values")
plt.legend(loc='upper right')

plt.subplot(3, 3, 5)
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points2")
plt.ylabel("Values2")
plt.legend(loc='upper right')

plt.subplot(3, 3, 9)
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points2")
plt.ylabel("Values2")
plt.legend(loc='upper right')

plt.show()
```

```python
plt.figure(figsize= (8, 3))

plt.subplot(1, 2, 1) # row=1, col=1, plot_no=1
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points")
plt.ylabel("Values")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, y, label='SINQ')
plt.plot(x, z, label='COSQ')
plt.title("SINQ & COSQ")
plt.xlabel("Data points2")
plt.ylabel("Values2")
plt.legend()
plt.subplots_adjust(wspace=0.8)
plt.grid()

plt.show()
```

```python
company = ['a', 'b', 'c', 'd']
revenue = [12, 34, 45, 23]
```

```python
plt.bar(company, revenue)
```

```python
plt.bar(company, revenue, color = ['red', 'blue', 'green', 'orange'])
```

```python
company = ['aljkgjfhgjhkjlkbhvgjhkl', 'bkhgfdfgufgu', 'c', 'd']
revenue = [12, 34, 45, 23]

plt.bar(company, revenue, color = ['blue', 'red', 'green', 'orange'])
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.barh(company, revenue, color = ['blue', 'red', 'green', 'orange'])
plt.title("Bar plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.hist(revenue)
plt.title("Histogram plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.pie(revenue)
plt.title("Pie plot")
plt.show()
```

```python
company = ['a', 'b', 'c', 'd']
revenue = [12, 34, 45, 23]
plt.pie(revenue, labels = company)
plt.title("Pie plot")
plt.show()
```

```python
# pie plot -> used to get the % of the distribution
plt.pie(revenue, labels = company, autopct = "%.2f%%")
plt.title("Pie plot")
plt.show()
```

```python
# Line plot -> used for trend over time
plt.plot(company, revenue)
plt.title("Line plot")
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.show()
```

```python
# box plot -> for detecting outliers
plt.boxplot(revenue)
plt.show()
```

```python
ls = [21, 34, 54, 23, 22]
np.mean(ls)
plt.boxplot(ls)
plt.show()
```
