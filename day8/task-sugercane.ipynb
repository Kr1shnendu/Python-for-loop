```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Production (Tons)'].str.replace('.', '')
```

```python
df['Production (Tons)'] = df['Production (Tons)'].str.replace('.', '')
df.sample(5)
```

```python
df['Production per Person (Kg)'] = df['Production per Person (Kg)'].str.replace(',', '')
df['Production per Person (Kg)'] = df['Production per Person (Kg)'].str.replace('.', '')

df.sample(5)
```

```python
df['Acreage (Hectare)'] = df['Acreage (Hectare)'].str.replace('.', '')
df.sample(5)
```

```python
df['Yield (Kg / Hectare)'] = df['Yield (Kg / Hectare)'].str.replace('.', '')
df['Yield (Kg / Hectare)'] = df['Yield (Kg / Hectare)'].str.replace(',', '')

df.sample(5)
```

```python
df.rename(
    columns={
        'Production (Tons)': 'Production_Tons',
        'Production per Person (Kg)': 'Production_per_Person_Kg',
        'Acreage (Hectare)': 'Acreage_Hectare',
        'Yield (Kg / Hectare)': 'Yield_Kg_per_Hectare'
    },
    inplace = True
)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.Acreage_Hectare.isnull()]
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```
