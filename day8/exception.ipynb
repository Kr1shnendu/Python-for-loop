```python
def safe_div():
    try:
        a = int(input("Enter a:"))
        b = int(input("Enter b:"))
        c = a / b
        print(a, '/', b, '=', c)
    except ZeroDivisionError:
        print("Divide by zero is not allowed")
        safe_div() #recursion
    except ValueError:
        print("Please enter only integer value")
        safe_div() #recursion

safe_div()
```
