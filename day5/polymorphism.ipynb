```python
class A:
    def __init__(self, a):
        self.a = a

    def __add__(self, other):
        return self.a + other.a
    
    def __sub__(self, other):
        return self.a - other.a
```

```python
o1 = A(10)
o2 = A(20)

print(o1 + o2)
print(o1 - o2)
```

```python
dir(A)
```

```python
class Vehicle:
    def __init__(self, vehicle_id, brand):
        self.vehicle_id = vehicle_id
        self.brand = brand

    def calculate_rent(self):
        print("Select vehicle...")

class Car(Vehicle):
    def __init__(self, vehicle_id, brand):
        super().__init__(vehicle_id, brand)

    def calculate_rent(self, days):
        print("Rent is 800 per day..")
        print("Your Rent is: ", days*800)

class Bike(Vehicle):
    def __init__(self, vehicle_id, brand):
        super().__init__(vehicle_id, brand)

    def calculate_rent(self, hours):
        print("Rent is 100 per hpur..")
        print("Your Rent is: ", hours*100)

class Truck(Vehicle):
    def __init__(self, vehicle_id, brand):
        super().__init__(vehicle_id, brand)

    def calculate_rent(self, distance):
        print("Rent is 75 per kilometer..")
        print("Your Rent is: ", distance*75)

```

```python
b = Car(2, "BMW")
b.calculate_rent(2)
```

```python
t = Truck(2, "Truck123")
t.calculate_rent(100)
```
