Merge and Concat

```python
data1 = {
    'name': ['A', 'B', 'C'],
    'age': [23, 45, 67],
    'salary': [100000, 200000, 360000]
}

data2 = {
    'name': ['A', 'Y', 'C'],
    'age': [23, 15, 67],
    'salary': [100000, 700000, 360000]
}

new_df1 = pd.DataFrame(data1)
new_df2 = pd.DataFrame(data2)
```

```python
new_df1
```

```python
new_df2
```

```python
pd.concat([new_df1, new_df2])
```

```python
pd.concat([new_df1, new_df2], ignore_index=True)
```

```python
pd.concat([new_df1, new_df2], ignore_index=True, axis=1)
```

```python
new_df = pd.concat([new_df1, new_df2], ignore_index=True)
new_df
```

```python
new_df.drop_duplicates(inplace=True)
new_df
```

```python
new_df1
```

```python
new_df2
```

```python
pd.merge(new_df1, new_df2, on='age', how='outer')
```
