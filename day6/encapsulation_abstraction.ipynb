```python
from abc import ABC, abstractmethod

class Bank(ABC):
    @abstractmethod
    def roi(self):
        pass

    @abstractmethod
    def loan(self):
        pass

class SBI(Bank):   #concrete class -> must implement all abstract methods
    def roi(self):
        print("7%")

    def loan(self):
        print("15%")
```

```python
obj = SBI()
obj.roi()
obj.loan()
```

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    def __init__(self, customer_name, date_of_birth):
        self.customer_name = customer_name
        self.date_of_birth = date_of_birth
        self.balance = 0

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

class SavingsAccount(BankAccount):
    min_balance = 3000

    def __init__(self, customer_name, date_of_birth):
        super().__init__(customer_name, date_of_birth)

    def deposit(self, amount):
        if (amount > 0):
            self.balance += amount
            print("Updated balance:", self.balance)
        else:
            print("Can't deposit amount:", amount)

    def withdraw(self, amount):
        if(self.balance - amount < self.min_balance):
            print(f"Can't withdraw {amount} \nAs balance is {self.balance} and min balance is {self.min_balance}")
        else:
            self.balance -= amount
            print(f"Successfully withdrawn {amount}, \nNow balance is {self.balance}")

class CurrentAccount(BankAccount):
    overdraft_limit = 10000

    def __init__(self, customer_name, date_of_birth):
        super().__init__(customer_name, date_of_birth)

    def deposit(self, amount):
        if (amount > 0):
            self.balance += amount
            print("Updated balance:", self.balance)
        else:
            print("Can't deposit amount:", amount)

    def withdraw(self, amount):
        if(self.balance - amount < -self.overdraft_limit):
            print(f"Can't withdraw {amount} \nAs balance is {self.balance} and overdraft limit is {self.overdraft_limit}")
        else:
            self.balance -= amount
            print(f"Successfully withdrawn {amount}, \nNow balance is {self.balance}")

        if(self.balance < 0):
            print(f"Pay {-self.balance} within a month")

```

```python
a = SavingsAccount("abc", '12-12-2012')
```

```python
a.deposit(30000)
```

```python
a.withdraw(10000)
```
