```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, roll, is_academic, is_sports):
        super().__init__(name, age)
        self.roll = roll
        self.is_academic = is_academic
        self.is_sports = is_sports

class Teacher(Person):
    def __init__(self, name, age, subject, experience):
        super().__init__(name, age)
        self.subject = subject
        self.experience = experience


class Academic:
    def __init__(self, course=None, marks=None):
        # if course==None:
        #     course = input("Enter Course name: ")
        # if marks==None:
        #     marks = int(input("Enter the marks: "))
        self.course = course
        self.marks = marks


class Sports:
    def __init__(self, sport_name=None , level=None):
        self.sport_name = sport_name
        self.level = level

class AllRounderStudent(Student, Academic, Sports):
    def __init__(self, name, age, roll, is_academic = False, is_sports = False,
                course=None, marks=None, sport_name=None , level=None
                ):
        super().__init__(name, age, roll, is_academic, is_sports)
        self.is_academic = is_academic
        self.is_sports = is_sports

        if self.is_academic:
            Academic.__init__(self, course, marks)

        if self.is_sports:
            Sports.__init__(self, sport_name, level)

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll:", self.roll)
        if self.is_academic:
            print("Marks:", self.marks)
            print("Course:", self.course)

        if self.is_sports:
            print("Sport name:", self.sport_name)
            print("Level:", self.level)
```

```python
s = AllRounderStudent("abc", 21, 4222636, is_academic=True, is_sports=True,
                        course="Science", marks=76, sport_name="Football", level="District")
```

```python
s.display()
```
