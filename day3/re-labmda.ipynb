```python
even_odd = lambda n: "Even" if n % 2 == 0 else "Odd"
print(even_odd(5))
```

```python
is_float = lambda x: "Float" if type(x) == float else "Not float"
print(is_float(10.5))
```

```python
add = lambda a, b, c: a+b+c
print(add(10, 20, 30))
```
