```python
class Bank:
    #static
    bank_name = "Bank of HIT"
    bank_loc = "Haldia"
    bank_contact = "100"
    bank_code = 235234

    def __init__(self, username, contact, age, balance):
        self.username = username
        self.contact = contact
        self.age = age
        self.balance = balance

    #object method
    def change_contact(self, contact):
        self.contact = contact

    #object method
    def display(self):
        print(self.username, self.contact, self.age, self.balance)

    def withdraw_money(self, amt):
        if(self.balance >= amt and amt > 0):
            self.balance -= amt
        print("Withdrawn money:", amt)

    def deposit_money(self, amt):
        if(amt > 0):
            self.balance += amt
        print("Deposited money:", amt)

    @classmethod
    def bank_details(self):
        print(self.bank_name,"  code:", self.bank_code)

    @classmethod
    def change_code(self, newcode):
        self.bank_code = newcode


# u1 = Bank()
# u1.username = "Abc"
# u1.contact = 63572572
# u1.age = 20

# u2 = Bank()
# u2.username = "User"
# u2.contact = 785868466
# u2.age = 30

# u3 = Bank()
# u3.username = "Xyz"
# u3.contact = 875646486
# u3.age = 25

u4 = Bank("ABC", 7566435457, 24, 10000)
print(u4.bank_name, u4.username, u4.contact, u4.age)

u5 = Bank("Gamma", 9786573457, 20, 100)
u5.display()
u5.change_contact(9999999999)
u5.display()
```

```python
Bank.bank_details()
Bank.change_code(555555)
Bank.bank_details()
```

```python
u6 = Bank("User6", 756453457, 20, 12000)
u6.display()
u6.withdraw_money(5000)
u6.display()
u6.deposit_money(20000)
u6.display()
```
