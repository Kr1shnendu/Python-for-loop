Comprehension

```python
list1 = [1, 2, 3, 4, 5, 6, 7, 8]
list2 = [i**2 for i in list1 if i % 2 == 0 ]
list2
```

```python
s = "this is a sentence to extract words"
l = [w for w in s.split(' ') if w[0].lower() in 'aeiou']
l
```

```python
list1 = [2, 2, 4, 6, 5, 7, 7, 8, 9, 20, 22, 22]
ans = {i for i in list1 if i > 5 }
ans
```

```python
ans = {i for i in range(1, 51) if i % 15 == 0}
ans
```

```python
dict1 = {i: i**3 for i in range(1, 6)}
dict1
```

```python
dict1 = {"a": 45, "b": 88, "c": 67, "d": 64, "e": 55}
dict2 = {i: dict1.get(i) for i in dict1.keys() if dict1.get(i) >= 60}
dict2
```
