```python
list1 = [1, 2, 3, 1, 2]

if list1[:2] == list1[-2:]:
    print("Valid slice")
```

```python
s = "madam"

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
s = "code9"
last = s[-1]

if last.isdigit():
    print("Ends with digit")
elif last.isalpha():
    print("Ends with character")
else:
    print("Ends with special symbol")
```

```python
s = "Python"

if s[0].isupper():
    print("Starts with uppercase")
elif s[-1].islower():
    print("Ends with lowercase")
else:
    print("Other case")
```

```python
lst = [4, 3, 1, 2]

mid = len(lst) // 2
first = sum(lst[:mid])
second = sum(lst[mid:])

if first > second:
    print("First half greater")
elif first == second:
    print("Equal halves")
else:
    print("Second half greater")
```

```python
s = "hello"

if len(s) >= 5:
    mid = len(s) // 2
    if s[mid].lower() in "aeiou":
        print("Middle character is a vowel")
    else:
        print("Middle character is not a vowel")
else:
    print("String too short")
```

```python
s = "ABZ"

if s.startswith("A"):
    if s.endswith("Z"):
        print("Valid AZ string")
    else:
        print("Starts with A but invalid end")
else:
    print("Invalid string")
```

```python
s = "12ab"

if s[:2].isdigit() and s[-2:].isalpha():
    print("Valid format")
else:
    print("Invalid format")
```

```python
lst = [5, 6, 1, 2]

if len(lst) % 2 == 0:
    if lst[:2] == lst[-2:]:
        print("Slices are same")
    else:
        print("Slices are different")
else:
    print("Odd length list")
```
