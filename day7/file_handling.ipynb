```python
with open('a.txt', 'r') as f:
    words = f.read().split()
    unique_words = set(words)
    print("Total unique words:", len(unique_words))
    print(unique_words)
```

```python
with open('a.txt', 'r') as f:
    words = f.read().split()
    d={}
    # for word in words:
    #     d[word] = 0
    # for word in words:
    #     d[word] += 1
    for word in words:
        if word not in d:
            d[word] = 1
        else:
            d[word] += 1
    print(d)
```

```python
import json

data = {
    'name': 'abc',
    'roll': 102,
    'age': 20
}

with open('json_1.json', 'w') as f:
    json.dump(data, f)
```

```python
with open('json_1.json', 'w+') as f:
    json.dump(data, f)
    f.seek(0)
    d = json.load(f)
    print(d)
```

```python
import csv

with open('data.csv', 'w', newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "Course"])
    writer.writerow(["Abc", "22", "CSE"])
```

```python
rows = [["Name", "Age", "Course"],["Abc", 22, "CSE"],["Hii", 20, "IT"]]

with open('data.csv', 'w', newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)
```

```python
with open('data.csv', 'w', newline="") as f:
    fieldnames = ["Name", "Age", "Course"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"Name": "Xyz", "Age": 21, "Course": "CSE"})
    writer.writerow({"Name": "Abc", "Age": 23, "Course": "IT"})
```

```python
with open("data.csv", "r") as f:
    reader = csv.DictReader(f)
    print(reader) #Object address
    for row in reader:
        print(row)
```
